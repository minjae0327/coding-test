```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def get_age(self):
        return self.age
    
    def get_name(self):
        return self.name
    
    def super_test(self):
        print("super_test")
```

## stack

```python
class Stack():
    def __init__(self):
        self.stack = []
        
    def size(self):
        return len(self.stack)
    
    def push(self, item):
        self.stack.append(item)
        
    def pop(self):
        if self.size() == 0:
            return -1
        else:
            return self.stack.pop()
        
    def top(self):
        if self.size() == 0:
            return -1
        else:
            return self.stack[-1]
    
    def find(self, item):
        try:
            return self.stack.index(item)
        except ValueError:
            return None

    def is_epmty(self):
        return self.size() == 0
```

## Queue

```python
class Queue():
    def __init__(self):
        self.queue = []
        
    def __len__(self):
        return len(self.queue)
    
    def enqueue(self, item):
        self.queue.append(item)
        
    def dequeue(self, item):
        if self.__len__ == 0:
            return None
        else:
            self.queue.pop(0)
```

## Linked List

```python
class LinkedList():
    def __init__(self):
        self.head = self.node(None)
        self.tail = self.node(None)
        self.head["next"] = self.tail
        
        
    def __len__(self):
        length = 0
        current_node = self.head
        while current_node["next"] != self.tail:
            length += 1
            current_node = current_node["next"]

        return length
        
        
    def node(self, data):
        return {"data": data, "next": None}
    
    
    def find_node(self, data):
        target_node = self.head["next"]
        while target_node != self.tail:
            if target_node["data"] == data:
                return target_node
            else:
                target_node = target_node["next"]
            
        return None
    
    
    def find_end_node(self):
        end_node = self.head
        if end_node["next"] is self.tail:
            return self.head
        
        while end_node["next"] != self.tail:
            end_node = end_node["next"]
            
        return end_node
    
    
    def add(self, data):
        new_node = self.node(data)
        end_node = self.find_end_node()
        end_node["next"] = new_node
        new_node["next"] = self.tail
        
            
    def remove(self, data):
        if self.head["next"] == self.tail:
            print("list is empty")
            return
        else:
            previous_node = self.head
            remove_node = self.head["next"]
            
            while remove_node != self.tail:
                if remove_node["data"] == data:
                    previous_node["next"] = remove_node["next"]
                    print("removed")
                    return
                else:
                    previous_node = remove_node
                    remove_node = remove_node["next"]
        
        print("There is no such data in the list")
        return
        
        
    def print_list(self):
        elements = []
        current_node = self.head
        
        while(1):
            if current_node == self.head:
                elements.append("head")
            elif current_node == self.tail:
                elements.append("tail")
                break
            else:
                elements.append(str(current_node["data"]))
            current_node = current_node["next"]
        
        return " -> ".join(elements)
```

```python
linked_list = LinkedList()
print(linked_list.__len__())
linked_list.add(1)
linked_list.add(2)
linked_list.add(3)
print(linked_list.__len__())
print("----------------------")
linked_list.print_list()

linked_list.remove(1)
print(linked_list.print_list())
```

## Graph

```python
import heapq
from collections import deque

class Graph:
    def __init__(self):
        self.graph = {}
        
        
    def add_vertex(self, vertex):
        if vertex not in self.graph:
            self.graph[vertex] = []
        
            
    def add_edge(self, start, end, weight=None, directed=False):
        if start not in self.graph:
            self.add_vertex(start)
        if end not in self.graph:
            self.add_vertex(end)
           
        self.graph[start].append([end, weight])
        if not directed:
            self.graph[end].append([start, weight])
            
    
    def display(self):
        for vertex in self.graph:
            print(f"{vertex}: {self.graph[vertex]}")
            
            
    def bfs(self, start):
        visited = set()
        visit_order = []
        have_to_visit = deque()
        have_to_visit.append(start)
        visited.add(start)
        
        while have_to_visit:
            vertex = have_to_visit.popleft()
            visit_order.append(vertex)
            
            for node in self.graph[vertex]:
                neighbor = node[0]
                
                if neighbor not in visited:
                    have_to_visit.append(neighbor)
                    visited.add(neighbor)
                    
        return visit_order
    
    
    def dfs(self, start):
        visited = set()
        visit_order = []
        have_to_visit = []
        have_to_visit.append(start)
        visited.add(start)
        
        while have_to_visit:
            vertex = have_to_visit.pop()
            visit_order.append(vertex)
            
            # 인접 노드를 역순으로 추가
            for node in reversed(self.graph.get(vertex, [])):
                neighbor = node[0]
                if neighbor not in visited:
                    have_to_visit.append(neighbor)
                    visited.add(neighbor)
                    
        return visit_order
    
    
    def dijkstra(self, start, end):
        distances = {vertex: float('inf') for vertex in self.graph}
        distances[start] = 0
        priority_queue = [(0, start)]
        previous_verticies = {vertex: None for vertex in self.graph}
        
        while priority_queue:
            current_distance, current_vertex = heapq.heappop(priority_queue)
            
            if current_vertex == end:
                break
        
            if current_distance > distances[current_vertex]:
                continue
            
            for neighbor, weight in self.graph[current_vertex]:
                if weight is None:
                    continue
                distance = current_distance + weight
                
                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    previous_verticies[neighbor] = current_vertex
                    heapq.heappush(priority_queue, (distance, neighbor))
                    
        shortest_distance = distances[end]
        if shortest_distance == float('inf'):
            return (float('inf'), [])
        
        path = []
        current = end
        while current is not None:
            path.insert(0, current)
            current = previous_verticies[current]
        
        return (shortest_distance, path)
     
     
    def floyd_warshall(self):
        vertices = list(self.graph.keys())
        distance = {u: {v: float('inf') for v in vertices} for u in vertices}
        next_vertex = {u: {v: None for v in vertices} for u in vertices}
       
        # 초기화: 간선 정보를 바탕으로 거리와 경로 설정
        for u in vertices:
            distance[u][u] = 0
            for edge in self.graph[u]:
                v, w = edge
                if w is not None:
                    distance[u][v] = min(distance[u][v], w)  # 다중 간선 시 최소 가중치 선택
                    next_vertex[u][v] = v
        
        # 플로이드-워셜 알고리즘 수행
        for k in vertices:
            for i in vertices:
                for j in vertices:
                    if distance[i][k] + distance[k][j] < distance[i][j]:
                        distance[i][j] = distance[i][k] + distance[k][j]
                        next_vertex[i][j] = next_vertex[i][k]
        
        return (distance, next_vertex)
        
        
    
    def display(self):
        for vertex in self.graph:
            print(f"{vertex}: {self.graph[vertex]}")
```

## Heap

```python
class MaxHeap():
    def __init__(self):
        self.max_heap = [None]
        
    
    def is_empty(self):
        return self.__len__() == 1
    
    
    def __len__(self):
        return len(self.max_heap)
            
    
    def insert_item(self, item):
        self.max_heap.append(item)
        self.insert_arrange()


    def remove_item(self):
        if self.is_empty():
            raise IndexError("The heap is empty")
        if self.__len__() == 2:
            return self.max_heap.pop()
        top_item = self.max_heap[1]
        self.max_heap[1] = self.max_heap.pop()
        self.remove_arrange()
        
        return top_item
        
    
    def insert_arrange(self):
        len_heap = self.__len__()
        
        if len_heap == 2:
            return
        
        current_index = len_heap - 1
        parent_index = current_index // 2
        
        
        while self.max_heap[current_index] > self.max_heap[parent_index]:
            self.max_heap[current_index], self.max_heap[parent_index] = \
                self.max_heap[parent_index], self.max_heap[current_index]
                
            current_index = parent_index
            parent_index = parent_index // 2
            if parent_index < 1:
                break
            
            
    def remove_arrange(self):
        current_index = 1
        len_heap = len(self.max_heap)
        
        while True:
            left_child_index = 2 * current_index
            right_child_index = 2 * current_index + 1
            largest = current_index
            
            # 왼쪽 자식이 존재하고 현재 노드보다 클 경우
            if left_child_index < len_heap and self.max_heap[left_child_index] > self.max_heap[largest]:
                largest = left_child_index
            # 오른쪽 자식이 존재하고 현재 가장 큰 노드보다 클 경우
            if right_child_index < len_heap and self.max_heap[right_child_index] > self.max_heap[largest]:
                largest = right_child_index
            # 가장 큰 노드가 현재 노드가 아니면 교환
            if largest != current_index:
                self.max_heap[current_index], self.max_heap[largest] = \
                    self.max_heap[largest], self.max_heap[current_index]
                current_index = largest
            else:
                break
            
```

```python
heap = MaxHeap()

heap.insert_item(4)
heap.insert_item(5)
heap.insert_item(6)
heap.insert_item(7)
heap.insert_item(9)

print(heap.max_heap)

print(heap.remove_item())
print(heap.max_heap)
```
